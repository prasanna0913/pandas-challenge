```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
#Calculating the total number of players
total_players = purchase_data["SN"].nunique()
total_players
#Putting into a data frame
Player_count= pd.DataFrame({"Total Player" : [total_players]})
Player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate unique items, average price, purchase count, and revenue
Unique_items = len(purchase_data["Item Name"].unique())
Unique_items
#Calculate average price
Average_price = purchase_data["Price"].mean()
Average_price
#Calculate purchase count
Number_of_pruchases = purchase_data["Purchase ID"].count()
Number_of_pruchases
#Calculate revenue
Total_Revenue = purchase_data["Price"].sum()
Total_Revenue
#putting all the of them in a dataframe
purchase_data_df = pd.DataFrame([{"Number of Unique Items": Unique_items, "Average Price": Average_price, 
                                 "Number of Purchases": Number_of_pruchases, "Total Revenue": Total_Revenue}])


#Applying the formats for currency
purchase_data_df["Average Price"] = purchase_data_df["Average Price"].map("${:,.2f}".format)
purchase_data_df["Total Revenue"] = purchase_data_df["Total Revenue"].map("${:,.2f}".format)

purchase_data_df.head()
```

```python
#Gender Demographics
```

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Count of male/female players and count of Other / Non-Disclosed 

male_players = purchase_data.loc[purchase_data["Gender"]=="Male", :]


count_male_players = male_players["SN"].nunique()
count_male_players

female_players = purchase_data.loc[purchase_data["Gender"]=="Female", :]


count_female_players = female_players["SN"].nunique()
count_female_players

other_players = purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed", :]


count_other_players = other_players["SN"].nunique()
count_other_players

Total_count = count_male_players +count_female_players + count_other_players
Total_count
#Calculating the percentage of male,female and other players
Percentage_of_male = (count_male_players/Total_count)*100
Percentage_of_female = (count_female_players/Total_count)*100
Percentage_of_other = (count_other_players/Total_count)*100
Percentage_of_other

#Putting all the data in the data frame
purchase_data_df = pd.DataFrame({
                         "Category": ["Male", "Female","Other/Non-Disclosed"], 
                         "Total_Count": [count_male_players, count_female_players, count_other_players],
                         "Percentage of Players": [Percentage_of_male, Percentage_of_female, Percentage_of_other]})
                                                                                        

#Applying format
purchase_data_df["Percentage of Players"] = purchase_data_df["Percentage of Players"].map("{0:.2f}%".format)
purchase_data_df["Percentage of Players"]

purchase_data_df.set_index("Category")
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# for male:
purchase_count_male = male_players["Purchase ID"].count()
avg_purchase_male = male_players["Price"].mean()
total_purchase_male = male_players["Price"].sum()
avg_per_person_male = total_purchase_male/count_male_players


# for female:
purchase_count_female = female_players["Purchase ID"].count()
avg_purchase_female = female_players["Price"].mean()
total_purchase_female = female_players["Price"].sum()
avg_per_person_female = total_purchase_female/count_female_players

# for other:
purchase_count_other = other_players["Purchase ID"].count()
avg_purchase_other = other_players["Price"].mean()
total_purchase_other = other_players["Price"].sum()
total_purchase_other = total_purchase_other/count_other_players

#Putting into a dataframe
summary_df= pd.DataFrame({"Gender" : ["Female", "Male", "Other/Non-Disclosed"],
                          "Purchase Count" : [purchase_count_female, purchase_count_male, purchase_count_other],
                          "Average Purchase Price" : [avg_purchase_female, avg_purchase_male, avg_purchase_other],
                          "Total Purchase Price" : [ total_purchase_female, total_purchase_male, total_purchase_other],
                          "Avg Total Purchase per Person" : [avg_per_person_female, avg_per_person_male, total_purchase_other]
                         
                         })        
                         
#Applying the formats for currency                       
summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:,.2f}".format)
summary_df["Total Purchase Price"] = summary_df["Total Purchase Price"].map("${:,.2f}".format)
summary_df["Avg Total Purchase per Person"] =summary_df["Avg Total Purchase per Person"].map("${:,.2f}".format)


summary_df.set_index("Gender")
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_df= purchase_data["Age"].min()
age_df= purchase_data["Age"].max()
age_df
# Establish bins for ages
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["less than 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins
purchase_data_age_group = pd.cut(purchase_data["Age"],bins, labels=group_names, include_lowest=True)
purchase_data_age_group

#place the "purchase_data_age_group" into a new column
purchase_data["Age Group"] = purchase_data_age_group
purchase_data
# groupby object upon "Age Group"
age_group_gb = purchase_data.groupby("Age Group")
age_group_gb
#count the total players by age
Number_gb = age_group_gb["SN"].nunique()
Number_gb
#Calculate percentae by age category
Percentage_gb = (Number_gb/total_players)*100
Percentage_gb
#Putting into a dataframe
age_demographics_df = pd.DataFrame({"Total Count" : Number_gb,
                                    "Percentage of Players" : Percentage_gb})                                
                                  
#Applying format
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{0:.2f}%".format)
age_demographics_df["Percentage of Players"]

age_demographics_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group purchase data by screen names
spenders_status = purchase_data.groupby("SN")
spenders_status
# Count the total purchases by name
purchase_count_spender = spenders_status["Purchase ID"].count()
purchase_count_spender
# Calculate the average purchase by name
avg_purchase_price_spender = spenders_status["Price"].mean()
avg_purchase_price_spender
#Calculate purchase total
Total_purchase_value = spenders_status["Price"].sum()
Total_purchase_value
#Crreating a dataframe
Top_spender_df = pd.DataFrame({"Purchase Count": purchase_count_spender,
                              "Average Purchase Price" : avg_purchase_price_spender,
                              "Total Purchase Value" : Total_purchase_value})
Top_spender_df
#Sort the total purchse value in desensing order
sortted_spenders = Top_spender_df.sort_values(["Total Purchase Value"], ascending=False)

sortted_spenders["Average Purchase Price"] = sortted_spenders["Average Purchase Price"].map("${:,.2f}".format)
sortted_spenders
sortted_spenders["Total Purchase Value"] = sortted_spenders["Total Purchase Value"].map("${:,.2f}".format)

sortted_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create a new data frame
popular_items = purchase_data[["Item ID", "Item Name", "Price"]]
popular_items
#group popular iems by item ID and item name.
popular_items_gb = popular_items.groupby(["Item ID", "Item Name"])
#count the number of times item has been purchase
purchase_count_item = popular_items_gb["Price"].count()

#cal total purchase value per item
purchase_value_item = popular_items_gb["Price"].sum()
purchase_value_item 

#cal average purchase value per item
avg_price_item = purchase_value_item/purchase_count_item
avg_price_item

#creating dataframe
most_popular_items_df = pd.DataFrame({"Purchase Count" : purchase_count_item,
                                "Item Price" : avg_price_item,
                                "Total Purchase Value" : purchase_value_item})
#sort purchase count column in descending order
sortted_items = most_popular_items_df.sort_values(["Purchase Count"], ascending=False)




#Format with currency$
sortted_items.head(5).style.format({"Item Price":"${:,.2f}", 
                                "Total Purchase Value":"${:,.2f}"})

```
